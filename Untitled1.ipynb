```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 4, 9, 16, 25]

plt.plot(x, y1, label="Linear", marker="o")
plt.plot(x, y2, label="Quadratic", marker="s")
plt.title("Multiple Plots Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # show labels
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 4, 9, 16, 25]

plt.subplot(1, 2, 1)  # (row, col, position)
plt.plot(x, y1, color="blue")
plt.title("Linear")

plt.subplot(1, 2, 2)
plt.plot(x, y2, color="green")
plt.title("Quadratic")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, marker="o")
plt.grid(True)  # grid  
plt.annotate("Peak", xy=(5, 25), xytext=(3, 20), arrowprops=dict(facecolor="black"))
plt.title("Annotation Example")
plt.show()
```

```python
# Install seaborn and matplotlib (only first time needed)
!pip install seaborn matplotlib

# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")   # - dataset

# 1. Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="time")
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()

# 2. Histogram
sns.histplot(tips["total_bill"], bins=20, kde=True, color="skyblue")
plt.title("Histogram: Total Bill")
plt.show()

# 3. Bar Plot
sns.barplot(x="day", y="total_bill", data=tips, ci=None, palette="viridis")
plt.title("Average Total Bill by Day")
plt.show()

# 4. Box Plot
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Box Plot: Total Bill by Day")
plt.show()

# 5. Violin Plot
sns.violinplot(x="day", y="total_bill", data=tips, palette="muted")
plt.title("Violin Plot: Total Bill by Day")
plt.show()

# 6. Heatmap (Correlation)
corr = tips.corr(numeric_only=True)  #  numeric  
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# 7. Pair Plot
sns.pairplot(tips, hue="sex")
plt.show()
```
